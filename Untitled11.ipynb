```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from fer import FER
from PIL import Image
```

```python
detector = FER(mtcnn=True)
```

```python
from google.colab import files
uploaded = files.upload()

for filename in uploaded.keys():
    image_path = filename
    break  # Only first image

# Load the image
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')
plt.title("Uploaded Image")
plt.show()
```

```python
# Detect emotions in the image
results = detector.detect_emotions(image_rgb)
```

```python
# Draw bounding boxes and labels
for result in results:
    (x, y, w, h) = result["box"]
    emotions = result["emotions"]
    dominant_emotion = max(emotions, key=emotions.get)

    cv2.rectangle(image_rgb, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(image_rgb, f'{dominant_emotion}', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX,
                0.9, (255, 0, 0), 2)

# Display result
plt.figure(figsize=(8, 6))
plt.imshow(image_rgb)
plt.axis('off')
plt.title("Emotion Detection Result")
plt.show()
```

```python
import cv2
import os

# Load Haar cascade for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Use uploaded video
video_path = next(iter(uploaded))
cap = cv2.VideoCapture(video_path)

# Output video settings
output_filename = 'output_with_faces.avi'
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = cap.get(cv2.CAP_PROP_FPS)

out = cv2.VideoWriter(output_filename, cv2.VideoWriter_fourcc(*'XVID'), fps, (frame_width, frame_height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    # Draw rectangles around faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

    # Write the frame with face boxes
    out.write(frame)

cap.release()
out.release()
print("Processing complete. Output saved as:", output_filename)
```
